```python
# Gradient Descent
import numpy as np
import matplotlib.pyplot as plt

x_data = [1.0, 2.0, 3.0 ]
y_data = [20.0, 40.0, 60.0]

#x를 넣어주면 y반환
def forward(x):
    return w * x #y_hat

def loss(x, y):
    y_pred = forward(x)
    return (y_pred - y) **2

#loss = (w * x - y) ^2
#dloos/dw = 2 * x * (w * x - y)
w = 0

def gradient(x,y ):
  return 2 * x * (w * x -y)



# 100 epoch
w = 0
lr = 0.01
for epoch in range(1000):
  l =0
  grad =0 
  for x, y in zip(x_data, y_data):
    grad += gradient(x,y)
    grad /= 3
    l += loss(x,y)

  w = w - lr * grad
  l /= 3
  print(f'epoch: {epoch}, w: {w}, l: {l}')

  
```

```python
import torch

x_data = [1.0, 2.0, 3.0 ]
y_data = [20.0, 40.0, 60.0]

w = torch.tensor([1.0], requires_grad = True)

# our model forward

#x를 넣어주면 y반환
def forward(x):
    return w * x #y_hat

def loss(x, y):
    y_pred = forward(x)
    return (y_pred - y) **2

for epoch in range(100):
  l =0 
  for x, y in zip(x_data, y_data):
    l += loss(x,y)
    
  l.backward()  # 대충 알아서 함
  w.data -= lr * w.grad.item()

  w.grad.data.zero_() # 0으로 초기화


  print(f'epoch: {epoch}, w: {w.data}')

```

```python
import torch
import torch.nn as nn
import torch.optim as optim

x_data = torch.Tensor([1.0, 2.0, 3.0 ])
y_data = torch.Tensor([20.0, 40.0, 60.0])

class Model(nn.Module):
  def __init__(self):
    super().__init__()
    self.w = torch.tensor([1.0], requires_grad = True)

  def forward(self, x):
   return self.w * x

loss = nn.MSELoss()
model = Model()
lr = 0.1

for epoch in range(20):
  y_pred = model(x_data)
  error = loss(y_pred, y_data)
  error.backward()
  
  model.w.data = model.w.data - lr * model.w.grad.item()
  model.w.grad.data.zero_()
  print(f'epoch: {epoch}, w: {model.w.data}')
```
